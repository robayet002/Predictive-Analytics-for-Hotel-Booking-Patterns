```python
df=df[df['adr']<5000]
```

# Data Analysis and Visualization

```python
canceled_perc= df['is_canceled'].value_counts(normalize=True)
canceled_perc
plt.figure(figsize=(5,3))
plt.title("Reservation status")
plt.bar(['Not canceled','Canceled'], df['is_canceled'].value_counts(), edgecolor='red')
```

```python
plt.figure(figsize=(8,5))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df)
plt.title("Reservation status in different hotel")
plt.xlabel("Hotel")
plt.ylabel("Reservation conuts")
```

```python
resort_hotel= df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True) 
```

```python
city_hotel= df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True) 
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(12,5))
plt.title("Average daily rate in resort and city hotel")
plt.plot(resort_hotel.index,resort_hotel['adr'],label="Resort Hotel")
plt.plot(city_hotel.index,city_hotel['adr'],label="City Hotel")
plt.legend()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(10,5))
plt.title( "Month wise cancelation status")
ax1=sns.countplot(x='month',hue='is_canceled',data=df)
plt.legend(["not canceled","canceled"])
plt.ylabel("number of reservations")
```

```python
plt.figure(figsize=(10,5))
plt.title("Adr per month")
grouped_df = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum()
grouped_df = grouped_df.reset_index()
sns.barplot(x='month', y='adr', data=grouped_df)
plt.show()
```

```python
canceled_data=df[df['is_canceled']==1]
top_10_country= canceled_data['country'].value_counts()[:10]
```

```python
plt.figure(figsize=(8,6))
plt.title("Top 10 country for canceling reservation",size=15)
plt.pie(top_10_country,labels=top_10_country.index,autopct="%.2f")
```

```python
df['market_segment'].value_counts(normalize=True)
```

```python
canceled_data['market_segment'].value_counts(normalize=True)
```

```python
canceled_df_adr=canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values("reservation_status_date",inplace=True)

not_canceled= df[df['is_canceled']==0]
not_canceled_df_adr=not_canceled.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values("reservation_status_date",inplace=True)
```

```python
canceled_df_adr= canceled_df_adr[(canceled_df_adr['reservation_status_date']>'2015-07') & (canceled_df_adr['reservation_status_date']<'2017-09')]
not_canceled_df_adr= not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date']>'2015-07') & (not_canceled_df_adr['reservation_status_date']<'2017-09')]


plt.figure(figsize=(12,5))
plt.title("Average daily rate for canceled and not canceled reservation")
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label="Canceled Reseravation")
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'],label="Not canceled Reseravation")
plt.legend()
```
